```python
import h5py
import matplotlib.pyplot as plt
import numpy as np

# Define file paths
signal_file = "input/test_input_signal.h5"
qcd_file = "input/test_input_QCD.h5"

# Function to load vae_lossY from h5 file
def load_ae_lossY(file_path):
    with h5py.File(file_path, "r") as f:
        ae_lossY = f["ae_lossY"][:]#Change to ae later
    return ae_lossY

# Load data
ae_lossY_signal = load_ae_lossY(signal_file)
ae_lossY_qcd = load_ae_lossY(qcd_file)

# Plot distributions
binLo = 0
binHi = 0.0002
nBins = 40

print(len(ae_lossY_qcd))
plt.figure(figsize=(8, 6))
plt.hist(vae_lossY_signal, bins=20, alpha=0.5, range=(binLo, binHi), label="Signal", density=True)
plt.hist(vae_lossY_qcd, bins=20, alpha=0.5, range=(binLo, binHi), label="QCD", density=True)
plt.xlabel("AE Loss Y")
plt.ylabel("Density")
plt.legend()
plt.show()
```
